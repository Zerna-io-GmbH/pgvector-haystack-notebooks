## Embedding Call with Haystack, PgVector and Ollama as Embedder

```python
from haystack import Document
from haystack_integrations.components.embedders.ollama import OllamaDocumentEmbedder
from dotenv import load_dotenv
load_dotenv()
```

```python
doc = Document(content="What do llamas say once you have thanked them? No probllama!")
document_embedder = OllamaDocumentEmbedder(url="http://192.168.178.82:11434/api/embeddings")
```

```python
result = document_embedder.run([doc])
print(result['documents'][0].embedding)
```
